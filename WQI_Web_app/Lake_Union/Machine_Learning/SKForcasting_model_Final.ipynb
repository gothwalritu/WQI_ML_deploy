```python
# Split data into train-test
# ==============================================================================
steps = 60
data_train = WQI_time_df[:-steps]
data_test  = WQI_time_df[-steps:]
print(f"Train dates : {data_train.index.min()} --- {data_train.index.max()}  (n={len(data_train)})")
print(f"Test dates  : {data_test.index.min()} --- {data_test.index.max()}  (n={len(data_test)})")
```

```python
fig, ax=plt.subplots(figsize=(10, 4))
plt.plot(data_train, label='Train')
plt.plot(data_test, label='Test')
plt.legend(loc='best')
```

```python

from sklearn.linear_model import LinearRegression
# Create and train forecaster
# ==============================================================================
#forecaster = ForecasterAutoreg(regressor = LinearRegression(), lags = 12)
forecaster = ForecasterAutoreg(regressor = LinearRegression(), lags = 12)
#forecaster = ForecasterAutoreg(regressor = RandomForestRegressor(random_state=123), lags = 6)
forecaster.fit(y=data_train['WQI'])
forecaster
```

```python
data_train
```

```python
# Predictions
# ==============================================================================
steps = 120
predictions = pd.DataFrame(forecaster.predict(steps=steps))
predictions.head(10)
```

```python
start = str(data_train.index[-1])
end = str("2030-12-01 00:00:00")
```

```python
predictions["date"] = pd.date_range(start, end, freq="MS")[0:steps]
```

```python
predictions.set_index("date", inplace=True)
```

```python
predictions.tail(12)
```

```python
fig, ax=plt.subplots(figsize=(10, 4))
plt.plot(data_train['WQI'], label='train')
plt.plot(data_test['WQI'], label='test')
plt.plot(predictions, label='predictions')
plt.legend(loc='best')
```

```python
# Test error
# ==============================================================================
error_mse = mean_squared_error(
                y_true = data_test['WQI'],
                y_pred = predictions
            )

print(f"Test error (mse): {error_mse}")
```
